```python
import pandas as pd
import re
import string
df = pd.read_csv('WikiPakistan.csv',header=None,names=['Data'])
df
```

```python
# Remove rows with missing or empty text
df = df.dropna(subset=['Data'])
df = df[df['Data'].str.strip() != '']
df['Data'] = df['Data'].astype(str)
df['Data'] = df['Data'].str.lower()
# Remove leading and trailing whitespace
df['Data'] = df['Data'].str.strip()

# Remove any extra whitespace within the text
df['Data'] = df['Data'].str.replace(r'\s+', ' ')
pattern = r'\(Urdu:.*?\)'
patterns = r'\[\d+\]'

# Apply re.sub to each element in the 'Data' column
df['Data'] = df['Data'].apply(lambda x: re.sub(patterns, '', x))
# Apply re.sub to each element in the 'Data' column
df['Data'] = df['Data'].apply(lambda x: re.sub(pattern, '', x))

#Remove punctuation
def remove_punctuation(text):
    translator = str.maketrans('', '', string.punctuation)
    cleaned_text = text.translate(translator)
    return cleaned_text

# Apply the function to the 'Data' column
df['Data'] = df['Data'].apply(remove_punctuation)
```

```python
df = df.reset_index()
```

```python
df.drop('index',inplace=True,axis=1)
```
